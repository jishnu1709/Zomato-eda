# VISUALIZATION

```python
# UNIVARIATE ANALYSIS - Analyze a single variable to understand its distribution or unique values.
```

```python
city_restaurant_counts = df_copy.groupby('city')['res_id'].count()
plt.figure(figsize=(12, 6))
plt.bar(city_restaurant_counts.index, city_restaurant_counts.values)
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Restaurant Concentration by City')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
city_restaurant_counts = df_copy.groupby('city')['res_id'].count().sort_values(ascending=False)

# Select the top 10 cities with the most restaurants
top_10_cities = city_restaurant_counts.head(10)

# Create a colorful bar chart
plt.figure(figsize=(12, 6))
colors = sns.color_palette("viridis", len(top_10_cities))  # Choose a vibrant color palette
bars = plt.bar(top_10_cities.index, top_10_cities.values, color=colors)
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Cities with the Most Restaurants')
plt.xticks(rotation=45, ha='right')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, yval, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Count plot for 'establishment'
plt.figure(figsize=(12, 6))
sns.countplot(data=df_copy, y='establishment', order=df_copy['establishment'].value_counts().index, palette='viridis')
plt.title('Distribution of Establishment Types')
plt.xlabel('Count')
plt.ylabel('Establishment')
plt.show()
```

```python
# Determine the most popular cuisines among the listed restaurants. make a colourfull graph

cuisine_counts = df_copy['cuisines'].str.split(', ', expand=True).stack().value_counts()

# Creating a colorful bar graph of the top 10 most popular cuisines
top_10_cuisines = cuisine_counts.head(10)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_10_cuisines.index, top_10_cuisines.values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightsalmon', 'lightblue', 'lightpink', 'lightyellow', 'lightseagreen', 'lavender'])
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Most Popular Cuisines')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by restaurant name and count the number of outlets
restaurant_counts = df_copy.groupby('name')['res_id'].count().sort_values(ascending=False)

# Select the top 10 restaurant chains
top_10_chains = restaurant_counts.head(10)

# Create a bar chart with a vibrant color palette
plt.figure(figsize=(12, 6))
colors = sns.color_palette("husl", len(top_10_chains))  # Choose a vibrant color palette
bars = plt.bar(top_10_chains.index, top_10_chains.values, color=colors)

plt.xlabel('Restaurant Chain')
plt.ylabel('Number of Outlets')
plt.title('Top 10 Restaurant Chains by Number of Outlets')
plt.xticks(rotation=45, ha='right')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, yval, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
df.rating_text.value_counts()
```

```python
data_ratingtext = {'rating_text': ['Excellent', 'Good', 'Average', 'Poor', 'Excellent', 'Very Good']}
df= pd.DataFrame(data_ratingtext)

# Count the occurrences of each rating_text
rating_counts = df['rating_text'].value_counts()

# Define a color palette
colors = sns.color_palette('Set3', n_colors=len(rating_counts))
explode = [0.1 if i == rating_counts.idxmax() else 0 for i in rating_counts.index]

plt.figure(figsize=(8,6))
plt.pie(
    rating_counts,
    labels=rating_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode)

plt.title('Rating Text Distribution', fontsize=16, fontweight='bold')
plt.legend(rating_counts.index, title="Ratings", loc='upper left', bbox_to_anchor=(1, 0.8))
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis - Analyze relationships between two variables.
```

```python
# Bar plot of average rating by establishment
plt.figure(figsize=(12, 6))
sns.barplot(data=df_copy, x='establishment', y='aggregate_rating', ci=None, palette='deep')
plt.title('Average Rating by Establishment Type')
plt.xlabel('Establishment Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for cost by city
plt.figure(figsize=(10,8))
sns.boxplot(data=df_copy, x='city', y='average_cost_for_two', palette='Set2')
plt.title('Cost for Two by City')
plt.xlabel('City')
plt.ylabel('Cost for Two')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for cost vs. rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_copy, x='average_cost_for_two', y='aggregate_rating', hue='price_range', palette='viridis')
plt.title('Cost for Two vs. Aggregate Rating')
plt.xlabel('Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Multivariate Analysis - Analyze combinations of variables to uncover patterns.
```

```python
# Stacked bar plot of rating text by city
rating_by_city = df_copy.groupby(['city','rating_text']).size().unstack(fill_value=0)
rating_by_city[['Excellent', 'Good', 'Average', 'Poor', 'Very Good']].plot(kind='bar', stacked=True, figsize=(15, 8), colormap='coolwarm')
plt.title('Rating Text Distribution by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for cost vs. rating by city

plt.figure(figsize=(12,10))
sns.scatterplot(data=df_copy, x='city', y='average_cost_for_two',hue='aggregate_rating', palette='dark')
plt.title('Cost for Two vs. Rating by City')
plt.xlabel('City')
plt.ylabel('Cost for Two')
plt.xticks(rotation=90)
plt.legend(title='Rating', loc='best')
plt.show()
```
